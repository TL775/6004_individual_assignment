```python
import pandas as pd

# 读取Excel文件
df = pd.read_excel('作图.xlsx')
df.head()
```

```python
import matplotlib.pyplot as plt

# 设置图表的宽度和高度
plt.figure(figsize=(10, 6))

# 定义颜色列表
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

# 绘制柱状图
df.plot(x='Unnamed: 0', kind='bar', rot=45, figsize=(10,6), color=colors)

# 设置图表标题和标签
plt.title('Model Performance by Feature Selection Method')
plt.xlabel('Feature Selection Method')
plt.ylabel('Accuracy')

# 显示图例
plt.legend(title='Model')

# 显示图表
plt.show()
```

```python
import seaborn as sns

# 将"Unnamed: 0"列设置为索引
df_indexed = df.set_index('Unnamed: 0')

# 设置热力图的大小
plt.figure(figsize=(10, 6))

# 绘制热力图
sns.heatmap(df_indexed, annot=True, cmap='coolwarm', fmt=".3f", linewidths=.5)

# 设置图表标题
plt.title('Model Performance by Feature Selection Method')

# 显示图表
plt.show()
```
